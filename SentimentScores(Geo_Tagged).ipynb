```
import pandas as pd
import csv

sentiment_scores = []
csvfile = "sentiment.csv"
df = pd.read_csv("march20_march21.csv")
#print(df.describe())

sentiment_scores.append(df.iloc[:,1].mean())

#print(sentiment_scores)

for i in range(20,29):
  df = pd.read_csv("march{}_march{}.csv".format(i,i+1))
  print(df.iloc[:,1].mean())
  sentiment_scores.append(df.iloc[:,1].mean())

df = pd.read_csv("march30_march31.csv")
print(df.iloc[:,1].mean())
sentiment_scores.append(df.iloc[:,1].mean())

df = pd.read_csv("march31_april1.csv")
print(df.iloc[:,1].mean())
sentiment_scores.append(df.iloc[:,1].mean())

for i in range(1,30):
  df = pd.read_csv("april{}_april{}.csv".format(i,i+1))
  print(df.iloc[:,1].mean())
  sentiment_scores.append(df.iloc[:,1].mean())

df = pd.read_csv("april30_may1.csv")
print(df.iloc[:,1].mean())
sentiment_scores.append(df.iloc[:,1].mean())

for i in range(1,31):
  df = pd.read_csv("may{}_may{}.csv".format(i,i+1))
  print(df.iloc[:,1].mean())
  sentiment_scores.append(df.iloc[:,1].mean())

df = pd.read_csv("may31_june1.csv")
print(df.iloc[:,1].mean())
sentiment_scores.append(df.iloc[:,1].mean())

for i in range(1,30):
  df = pd.read_csv("june{}_june{}.csv".format(i,i+1))
  print(df.iloc[:,1].mean())
  sentiment_scores.append(df.iloc[:,1].mean())

df = pd.read_csv("june30_july1.csv")
print(df.iloc[:,1].mean())
sentiment_scores.append(df.iloc[:,1].mean())

for i in range(1,19):
  df = pd.read_csv("july{}_july{}.csv".format(i,i+1))
  print(df.iloc[:,1].mean())
  sentiment_scores.append(df.iloc[:,1].mean())

with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    for val in sentiment_scores:
        writer.writerow([val])    
```
